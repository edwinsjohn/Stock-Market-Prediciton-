```python
from tensorflow.keras.models import load_model
new_model=load_model('models/sentiment_analysis_stock.h5')
```

```python
#modules for Keras gym and tensorflow
from __future__ import division, print_function
from gensim import models
from keras.callbacks import ModelCheckpoint
from keras.layers import Dense, Dropout, Reshape, Flatten, concatenate, Input, Embedding
from keras.layers.recurrent import LSTM
from keras.models import Sequential
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.models import Model
from sklearn.model_selection import train_test_split
from nltk import word_tokenize, WordNetLemmatizer
import numpy as np
import pandas as pd
import os
import collections
import re
import string
```

```python
#converting tokens from Uppercase to Lowercase
def lower_token(tokens): 
    return [w.lower() for w in tokens]    
```

```python
from nltk.corpus import stopwords
# nltk.download('stopwords')
stoplist = stopwords.words('english')
def remove_stop_words(tokens):
    return [word for word in tokens if word not in stoplist]
```

```python
inputs=input("Enter the news:")
```

```python
import pickle
with open("TRAINING_VOCAB", "rb") as fp:   #Pickling
    TRAINING_VOCAB=pickle.load(fp)
with open("TEXT_DATA", "rb") as fp:   #Pickling
    TEXT_DATA=pickle.load(fp)    
MAX_SEQUENCE_LENGTH = 50
EMBEDDING_DIM = 300    
TRAINING_VOCAB
tokenizer = Tokenizer(num_words=len(TRAINING_VOCAB), lower=True, char_level=False)
tokenizer.fit_on_texts(TEXT_DATA)
TEXT_DATA
```

```python
input_un_tokens = [word_tokenize(inputs)]
# print(input_un_tokens)
input_tokens=input_un_tokens[0]
remove_sym=[';',':','!','@','#','$','%',',','.','^','&','*','(',')','/','?','`','...','..','--','---','~',"+","=",'-',"}","{"]
for i in input_tokens:
    if i in remove_sym:
        input_tokens.remove(i)
# print(input_tokens)
new_tokens=[]
new_tokens.append(input_tokens)
lower_input_tokens = [lower_token(in_token) for in_token in new_tokens] 
print(lower_input_tokens)
filtered_words = [remove_stop_words(sen) for sen in lower_input_tokens] 
sentance = [' '.join(sen) for sen in filtered_words]
print(sentance)
input_sequences = tokenizer.texts_to_sequences(sentance)
cnn_data = pad_sequences(input_sequences, maxlen=MAX_SEQUENCE_LENGTH)
input_sequences
cnn_data
# tokenizer
```

```python
labels = [1, 0,.5]
p=new_model.predict(cnn_data)
print(labels[np.argmax(p)])
```
